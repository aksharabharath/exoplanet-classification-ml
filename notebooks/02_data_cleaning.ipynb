# 02 - Data Cleaning

This notebook prepares a machine learning-ready dataset to study whether stellar properties can predict the presence of gas giant planets.


```python
import pandas as pd

df = pd.read_csv("../data/processed/exoplanets_clean.csv")
df.head()
```

```python
cols = [
    "pl_rade",   # needed to define gas giant label
    "st_teff",   # stellar temperature
    "st_rad",    # stellar radius
    "st_met",    # metallicity (IMPORTANT for your hypothesis)
    "st_age"     # stellar age
]

df_model = df[cols].copy()
df_model.head()
```

```python
df_model.isna().sum()
```

```python
df_model = df_model.dropna()
df_model.shape
```

```python
df_model["is_gas_giant"] = (df_model["pl_rade"] > 6).astype(int)
df_model.head()
```

```python
df_model["is_gas_giant"].value_counts()
```

```python
X = df_model.drop(columns=["pl_rade", "is_gas_giant"])
y = df_model["is_gas_giant"]

X.head()
```

```python
df_model.to_csv("../data/processed/ml_ready_dataset.csv", index=False)
```
