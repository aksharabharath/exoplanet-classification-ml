## Key Observation

All stellar features show strong overlap between gas giant and non-gas giant systems, indicating that no single stellar property independently separates the two classes.

However, gas giant host systems exhibit a wider distribution range, suggesting higher variability in stellar environments where gas giants are found.

This implies that gas giant formation is likely influenced by a combination of multiple stellar factors rather than a single dominant feature.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/ml_ready_dataset.csv")
df.head()
```

```python
gas = df[df["is_gas_giant"] == 1]
non_gas = df[df["is_gas_giant"] == 0]
```

```python
plt.hist(non_gas["st_teff"], bins=50, alpha=0.6, label="Not Gas Giant")
plt.hist(gas["st_teff"], bins=50, alpha=0.6, label="Gas Giant")

plt.legend()
plt.title("Stellar Temperature Distribution")
plt.xlabel("Temperature (K)")
plt.show()
```

```python
plt.hist(non_gas["st_rad"], bins=50, alpha=0.6, label="Not Gas Giant")
plt.hist(gas["st_rad"], bins=50, alpha=0.6, label="Gas Giant")

plt.legend()
plt.title("Stellar Radius Distribution")
plt.show()
```

```python
plt.hist(non_gas["st_met"], bins=50, alpha=0.6, label="Not Gas Giant")
plt.hist(gas["st_met"], bins=50, alpha=0.6, label="Gas Giant")

plt.legend()
plt.title("Stellar Metallicity Distribution")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("../data/processed/exoplanets_clean.csv")

df.head()
```

```python
df["is_gas_giant"] = (df["pl_rade"] > 6).astype(int)
```

```python
df["is_gas_giant"].value_counts()
```

```python
def classify_planet(radius):
    if radius < 1.25:
        return "Terrestrial"
    elif radius < 2:
        return "Super-Earth"
    elif radius < 6:
        return "Mini-Neptune"
    else:
        return "Gas Giant"

df["planet_type"] = df["pl_rade"].apply(classify_planet)
```

```python
plt.figure(figsize=(6,4))

df["planet_type"].value_counts().plot(kind="bar")

plt.title("Planet Type Distribution")
plt.xlabel("Planet Type")
plt.ylabel("Count")
plt.xticks(rotation=30)

plt.show()
```

```python
plt.figure(figsize=(7,4))

sns.boxplot(x="is_gas_giant", y="st_teff", data=df)

plt.title("Stellar Temperature vs Gas Giant Presence")
plt.xlabel("Gas Giant (0 = No, 1 = Yes)")
plt.ylabel("Stellar Temperature (K)")

plt.show()
```

```python
plt.figure(figsize=(7,4))

sns.boxplot(x="is_gas_giant", y="st_rad", data=df)

plt.title("Stellar Radius vs Gas Giant Presence")
plt.xlabel("Gas Giant (0 = No, 1 = Yes)")
plt.ylabel("Stellar Radius")

plt.show()
```

```python
plt.figure(figsize=(7,4))

sns.boxplot(x="is_gas_giant", y="st_met", data=df)

plt.title("Metallicity vs Gas Giant Presence")
plt.xlabel("Gas Giant (0 = No, 1 = Yes)")
plt.ylabel("Metallicity")

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")

plt.title("Feature Correlation Heatmap")

plt.show()
```
